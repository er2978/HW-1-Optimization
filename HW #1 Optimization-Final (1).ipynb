```python
import matplotlib.pyplot as plt
from math import cos, pi, sin
import numpy as np

def f(x,y):
    return (1-cos(pi*x/2))*y**2
xi, yi = np.meshgrid(np.linspace(-1,1,1000), np.linspace(-1,1,1000))
f_eval = np.zeros(xi.shape)
mask = []
for i in range(1000):
    for j in range(1000):
        f_eval[i,j] = f(xi[i,j],yi[i,j])
        
plt.figure(figsize = (6,5))
plt.axis('equal')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Experimenting With Gradient Descent Cost Function")
plt.contourf(xi, yi, f_eval)
plt.colorbar()
```

```python
def grad_f(xi, yi):
    return np.array([pi*yi**2*sin(pi*xi/2)/2,2*yi*(1-cos(pi*xi/2))])

r0 = [tuple(np.random.uniform(-1,1,2))]
grad_r0 = grad_f(*r0[0])

print(r0, grad_r0)


for _ in range(1):
    r0 = [tuple(np.random.uniform(-1, 1, 2))]
    for i in range(10):
          r0.append(tuple(np.subtract(r0[-1], grad_f(*r0[-1]))))     
    r = np.array(r0)
    plt.plot(*r.T, c='w', linewidth=0.4, markersize=1, marker = '.')
    im = plt.contourf(xi, yi, f_eval, levels = 10)
    plt.title("Experimenting With Gradient Descent 10 Steps")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.colorbar(im)
    plt.show(block = False)
    plt.figure(figsize = (6,5))
    plt.axis('equal')
   
```
