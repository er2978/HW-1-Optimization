```python
import matplotlib.pyplot as plt
from math import cos, pi, sin
import numpy as np

def f(x,y):
    return (1-cos(pi*x/2))*y**2
xi, yi = np.meshgrid(np.linspace(-1,1,1000), np.linspace(-1,1,1000))
f_eval = np.zeros(xi.shape)
mask = []
for i in range(1000):
    for j in range(1000):
        f_eval[i,j] = f(xi[i,j],yi[i,j])
        
plt.figure(figsize = (6,5))
plt.axis('equal')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Experimenting With Gradient Descent Cost Function")
plt.contourf(xi, yi, f_eval)
plt.colorbar()
```
